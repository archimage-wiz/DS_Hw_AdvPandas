### Задание 1


```python
import pandas as pd
```

```python
vl = pd.read_csv("visit_log.csv", sep=";")
```

```python
vl.head(10)
```

```python
vl_res = vl.copy()

vl_res["source_type"] = vl_res["traffic_source"]
filter_condition = (vl_res["traffic_source"].str.lower() == "paid") | (
    vl_res["traffic_source"].str.lower() == "email"
)
vl_res.loc[
    (vl_res["traffic_source"].str.lower() == "yandex")
    | (vl_res["traffic_source"].str.lower() == "google"),
    "source_type",
] = "organic"
vl_res.loc[
    filter_condition & (vl_res["region"] == "Russia"),
    "source_type",
] = "ad"
vl_res.loc[
    filter_condition & (vl_res["region"] != "Russia"),
    "source_type",
] = "other"

vl_res.tail(10)
```

### Задание 3


```python
ratings_data = pd.read_csv("ratings.csv")

ratings_data.head(10)
```

```python
res = (
    ratings_data.groupby("userId")
    .filter(lambda x: x["timestamp"].count() > 100)
    .groupby("userId")
    .agg(
        {
            "timestamp": lambda x: max(x) - min(x),
        }
    )
    .reset_index()
    .rename(columns={"timestamp": "time_diff"})
)
res.head(10)
```

```python
res["time_diff"].mean()
```

```python
d1 = rzd.merge(auto, how="outer", left_on="client_id", right_on="client_id")
d2 = d1.merge(air, how="outer", left_on="client_id", right_on="client_id")
d2.fillna(0, inplace=True)
d2
```
